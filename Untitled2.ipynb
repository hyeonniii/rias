```python
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
processed_data=data.iloc[:,1:11]
```

```python
distortions = []
K= range(1,10)
for k in K:
  kmeanModel= KMeans(n_clusters=k)
  kmeanModel.fit(data.iloc[:,1:9])
  distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K,distortions,'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.show()
```

```python
kmeanModel=KMeans(n_clusters=3)
kmeanModel.fit(data.iloc[:,1:9])
```

```python
result_by_sklearn = data.iloc[:,1:9]
result_by_sklearn["cluster"]=kmeanModel.labels_
result_by_sklearn.head()
```

```python
kmeanModel=KMeans(n_clusters=2)
kmeanModel.fit(data.iloc[:,1:10])

result_by_sklearn = data.iloc[:,1:10]
result_by_sklearn["cluster"]=kmeanModel.labels_
result_by_sklearn.head()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
score_samples=silhouette_samples(processed_data,result_by_sklearn["cluster"])
print(score_samples.shape)
```

```python
result_by_sklearn['silhouette_coeff']=score_samples
```

```python
average_score=silhouette_score(processed_data,result_by_sklearn['cluster'])
print('analysis score',format(average_score))
result_by_sklearn.head()
```
